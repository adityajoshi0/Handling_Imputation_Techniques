Dealing with Missing values

Imputation Techniques
-> Mean, Median, Mode

```python
data = pd.read_csv('/content/CompanyProfit.csv')
data1 = pd.read_csv('/content/CompanyProfit_new.csv')
data2 = pd.read_csv('/content/CompanySuccess.csv')
```

```python
data.head(10)
```

```python
# Checking empty values
data.isnull().sum()
```

```python
import seaborn as sns
```

```python
# Distribution plot to identify Which Technique to use
sns.distplot(data['Profit ($)'])
```

```python
# Fillana Method Using Mean
col = data['Profit ($)']
col = col.fillna(col.mean(), inplace = True)
```

```python
# After fillana method
data.isnull().sum()
```

```python
data.head(20)
```

Median techniques


```python
data1.isnull().sum()
```

```python
# Distribution Plot to identify Which Technique to use

sns.distplot(data1['Profit ($)'])
```

```python
# # Fillna Method Using Median

col = data1['Profit ($)']

col = col.fillna(col.median(), inplace = True)
```

```python
# After Fillna Method

data1.isnull().sum()
```

```python
# Mean - For Normal Distribution
# Median - Right or Left Skewed Data   (Not Normal Distribution)
```

Mode Techniques

```python
data2.head(20)
```

```python
#chechking null values
data2.isnull().sum()
```

```python
# fillana Method Using Mode
col = data2['Success']
col = col.fillna(col.mode()[0], inplace = True)
```
